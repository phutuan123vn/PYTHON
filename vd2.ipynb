```python
hist,bin_edge=np.histogram(data,bins=np.arange(256),density=True)
```

```python

lk=hist.cumsum()
lk
```

```python
uT=sum(np.multiply(np.arange(255),hist))
ut=np.multiply(np.arange(255),hist)
ut=ut.cumsum()
```

```python
a=np.subtract(lk*uT,ut)
a=a**2
b=lk*(1-lk)
```

```python
c=a[:253]/b[:253]
t=np.unravel_index(np.argmax(c,axis=0),c.shape[0])
```
